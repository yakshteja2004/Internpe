**Breast cancer classification with a simple neural network(NN)**

Importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data collection & Processing

```python
# load the dataset
breast_cancer=sklearn.datasets.load_breast_cancer()
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer.target
```

Separating the features and target

```python
X= data_frame.drop(columns='label', axis='1')
Y= data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Spliting the data into training the data and testing the data

```python
X_train,Y_train,X_test,Y_test=train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

Building neural network

download.png

```python
# importing tensorflow as ts
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers neural network

model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network
from sklearn.preprocessing import StandardScaler

# Standardize the training features
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)

# Due to an incorrect variable assignment in a previous cell (CamAbeG25TX9),
# the actual training labels are currently stored in the variable named 'X_test'.
# We use this variable as the target for training.
correct_Y_train = X_test

history = model.fit(X_train_std, correct_Y_train, validation_split=0.1, epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```

Accuracy on the model on test data

```python
# Standardize the actual test features using the scaler fitted on the training data.
# Note: Due to an incorrect variable assignment in cell 'CamAbeG25TX9',
# the actual test features are currently stored in the 'Y_train' variable.
X_test_std = scaler.transform(Y_train)

# Evaluate the model using the standardized test features and the correct test labels.
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```

Building the predictive system

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
